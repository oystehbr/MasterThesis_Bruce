```python
best = []
start_scenario = 12
for idx, score_idx in enumerate(np_scores_max_idx):

    pd_stats = pd.read_csv(f'statistics_latex_format/scenario{start_scenario + score_idx}.txt', delimiter='&', names=['maps', 'dumper_f', 'fuel_f', 'time_f', 'score_f', 'dumper_1', 'fuel_1', 'time_1', 'score_1', 'fuel_2', 'time_2'])
    best.append(list(pd_stats.iloc[idx*2].values[1:]))

```

```python
best_framework_dumpers = np.array(best)[:, 0]
best_framework_fuel = np.array(best)[:, 1]
best_framework_time = np.array(best)[:, 2]
best_framework_scores = np.array(best)[:, 3]
benchmark1_dumpers = np.array(best)[:, 4]
benchmark1_fuel = np.array(best)[:, 5]
benchmark1_time = np.array(best)[:, 6]
benchmark1_scores = np.array(best)[:, 7]

print(np.mean(best_framework_fuel)/np.mean(benchmark1_fuel))
print(np.mean(best_framework_time)/np.mean(benchmark1_time))
```
